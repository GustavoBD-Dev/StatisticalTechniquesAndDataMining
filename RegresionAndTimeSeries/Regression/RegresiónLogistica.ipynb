Se muestran los datos de una muestra de 53 pacientes con cáncer de próstata en los que se mide la Edad, Nivel de ácido que mide la extensión del tumor, grado de gravedad del tumor, Etapa en la que se encuentra, Resultados de una radiografía y, cuándo se ha detectado al intervenir quirúrgicamente que el cáncer se ha extendido a los nodos linfáticos. El objetivo es ajustar un modelo que permita predecir cuándo el cáncer se extiende a los nodos linfáticos (o no) sin necesidad de intervención quirúrgica.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import subplots

pd.options.mode.chained_assignment = None

muestra = pd.read_csv('glm.csv')
muestra.head()
```

```python
# Convertir la columna Y a binaria
muestra['NODOS'] = muestra['NODOS'].map({'Neg': 0, 'Pos': 1})
```

```python
muestra.head()
```

```python
X = muestra.drop(columns = 'ID')
X.head()
```

```python
X = X.drop(columns = 'NODOS')
X.head()
```

```python
import statsmodels.api as sm
#  estimación y validación de regersión
# Iniciamos el proceso de la determinación del modelo de regresion lineal
X = pd.DataFrame({'beta0' : np.ones(muestra.shape[0]), 'beta1' : muestra ['RADIOGRA'], 'beta2' : muestra['ETAPA'],
                  'beta3' : muestra ['GRADO'], 'beta4' : muestra ['EDAD'], 'beta5' : muestra ['ACIDO'] })
#diccionario Lave : valor
X.head()
```

```python
Y = muestra['NODOS']
Y.head()
```

```python
# Especificamos el modelo de regresión
mod = sm.GLM(Y, X, family = sm.families.Binomial())

# Ajustamos el modelo
resul = mod.fit()

# Imprimimos los resultados
print(resul.summary())
```

1. Deviance: es la devianza del modelo.
2. Log-Likelihood: es la log verosimilitud
3. Pearson-Chi2: ji cuadrado
4. Pseudo Qsqu: R^2 de Cox-Snell

```python
# Predicciones con intervalo de confianza
# ==============================================================================
pronostico = resul.predict(exog = X)

# Clasificación predicha
# ==============================================================================
clasificacion = np.where(pronostico<0.5, 0, 1)
clasificacion
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
# División de los datos en train y test
# ==============================================================================
X = muestra.drop(columns = 'NODOS')
y = muestra['NODOS']

X_train, X_test, y_train, y_test = train_test_split(
                                        X,
                                        y.values.reshape(-1,1),
                                        train_size   = 0.8,
                                        random_state = 1234,
                                        shuffle      = True
                                    )

# Creación del modelo utilizando matrices como en scikitlearn
# ==============================================================================
# A la matriz de predictores se le tiene que añadir una columna de 1s para el intercept del modelo
X_train = sm.add_constant(X_train, prepend=True)
modelo = sm.Logit(endog=y_train, exog=X_train,)
modelo = modelo.fit()
print(modelo.summary())
```

```python
#from sklearn.metrics import confusion_matrix #, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
# Accuracy de test del modelo
# ==============================================================================
X_test = sm.add_constant(X_test, prepend=True)
predicciones = modelo.predict(exog = X_test)
clasificacion = np.where(predicciones<0.5, 0, 1)
accuracy = accuracy_score(
            y_true    = y_test,
            y_pred    = clasificacion,
            normalize = True
           )
print("")
print(f"El accuracy de test es: {100*accuracy}%")
```

```python
#Matriz de confusión de las predicciones de test
# ==============================================================================
confusion_matrix = pd.crosstab(
    y_test.ravel(),
    clasificacion,
    rownames=['Real'],
    colnames=['Predicción']
)
confusion_matrix
```
