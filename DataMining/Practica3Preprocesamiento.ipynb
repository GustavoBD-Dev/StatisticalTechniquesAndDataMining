### Preprocesamiento
Supongamos que los datos para el análisis incluyen el atributo de Edad. Los valores de la edad para las tuplas de datos son (en orden creciente): 12, 15, 16, 16, 19,19, 20, 20, 20, 21, 22, 22, 25, 25, 25, 25, 30, 33, 34, 35, 35, 35, 35, 36, 36,40, 45, 46, 51, 70, 72, 77

* ¿Cuál es la media de los datos? ¿Cuál es la mediana?
* ¿Cuál es la moda de los datos?
* Utilice la normalización min-max para transformar el valor de 44 de la antigúedad en el intervalo [0.0, 1.0]
* Trazar un diagrama equi-distancia de longitud 8.

```python
# Declaramos los datos
data = np.array([12, 15, 16, 16, 19,19, 20, 20, 20, 21,
                 22, 22, 25, 25, 25, 25, 30, 33, 34, 35,
                 35, 35, 35, 36, 36,40, 45, 46, 51, 70, 72, 77])

# Calculo de los valores
media = np.mean(data)
mediana = np.median(data)
moda = np.argmax(np.bincount(data))
normalizacion = (44 - np.min(data)) / (np.max(data) - np.min(data))

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
print(f"El valor normalizado para 44 es: {normalizacion}")
```

```python
# Definir el rango de los bins
min_val = np.min(data)
max_val = np.max(data)
bin_width = 8
num_bins = int((max_val - min_val) / bin_width) + 1

# Crear los bins con intervalos
bins = np.arange(min_val, max_val + bin_width, bin_width)

# Crear el histograma
plt.hist(data, bins=bins, edgecolor='black')

# Etiquetas de los ejes y título
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Histograma de Equi-Distancia (Longitud 8)")

# Mostrar los intervalos en el eje x
plt.xticks(bins)

# Mostrar el gráfico
plt.show()
```

### Preprocesamiento

Supongamos que un hospital probó los datos de la edad y grasa corporal para 18 adultos seleccionados al azar con el siguiente resultado:

a) Calcule la media, la mediana y la desviacion estandar de la edad y % Grasa.

b) Grafica el diagrama de dispersión

c) Normaliza los datos de Edad y % Grasa

d) Calcula y explica el coeficiente de correlación de Pearson

Screenshot 2024-09-22 at 6.24.06 a.m..png

```python
# Definimos los valores de las variables
edad = np.array([22, 23, 27, 27, 39, 42, 47, 49, 50,
                 52, 54, 54, 56, 57, 59, 59, 62, 66])
grasa = np.array([9.5, 26.5, 7.7, 18.8, 31.4, 25.9, 27.5, 27.2, 31.3,
                  34.5, 42.5, 28.9, 33.5, 30.2, 34.1, 33.3, 41.4, 35.7])

# Calculamos las estadísticas para cada variable
print("Estadisticas para la variable EDAD")
media_edad = np.mean(edad)
mediana_edad = np.median(edad)
desviacion_edad = np.std(edad)
print(f"Media: {round(media_edad,2)}")
print(f"Mediana: {mediana_edad}")
print(f"Desviacion estandar: {round(desviacion_edad,2)}\n")

print("Estadisticas para la variable %GRASA")
media_grasa = np.mean(grasa)
mediana_grasa = np.median(grasa)
desviacion_grasa = np.std(grasa)
print(f"Media: {round(media_grasa,2)}")
print(f"Mediana: {mediana_grasa}")
print(f"Desviación estandar: {round(desviacion_grasa,2)}")
```

```python
# Diagrama de dispersión de las variables
plt.scatter(edad, grasa)
plt.xlabel("Edad")
plt.ylabel("% Grasa")
plt.title("Diagrama de dispersión de Grasa vs Edad")
plt.show()
```

```python
# Generamos un Dataframe con las variables originales
data = {'Edad': edad, '%Grasa': grasa}
df = pd.DataFrame(data)

# Calculamos los datos normalizados y los agregamos a columnas nuevas del dataframe
df['Edad_norm'] = round((df['Edad'] - df['Edad'].min()) / (df['Edad'].max() - df['Edad'].min()),3)
df['%Grasa_norm'] = round((df['%Grasa'] - df['%Grasa'].min()) / (df['%Grasa'].max() - df['%Grasa'].min()),3)

df
```

```python
# Tamaño de la variable
fig = plt.subplots(1, sharex='col', figsize=(6, 3))
labels = ['Edad','%Grasa'] # etiqueta de las columnas
df_corr = df[['Edad', '%Grasa']] # solo variables originales
# Calculo del coeficiente de correlacion
matrix = pd.DataFrame(df_corr.corr('pearson').values, index=labels, columns=labels)
# Mapa de calor
sns.heatmap(matrix, annot=True, fmt='.3f', cmap=sns.cubehelix_palette(rot=-.1))
plt.title('Coeficiente de correlación de Pearson')
plt.show()
```

El valor en la celda que cruza "Edad" y "%Grasa" (y viceversa) es de **0.815**. Este valor indica una correlación positiva fuerte entre ambas variables. Esto significa que, a medida que aumenta la edad, tiende a aumentar también el porcentaje de grasa corporal.

### Práctica 3 Distancias


Cálculo de las distancias, atributos numéricos.

Sean la tuplas: $X_1=(7,2)$ y $X_2=(4,8)$

Encuentre la distancia utilizando:

a) Distancia maxima

b) Distancia de manhatan

c) Distancia Euclidiana

Expresar graficamente los resultados

```python
x_1 = np.array([7, 2])
x_2 = np.array([4, 8])

distancia_maxima = np.max(np.abs(x_1 - x_2))
distancia_manhatan = np.sum(np.abs(x_1 - x_2))
distancia_euclidiana = np.sqrt(np.sum((x_1 - x_2) ** 2))

print("Distancia maxima:", distancia_maxima)
print("Distancia de Manhattan:", distancia_manhatan)
print("Distancia Euclidiana:", str(round(distancia_euclidiana, 2)))
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(x_1[0], x_1[1], color='red', label='X1 (7,2)')
plt.scatter(x_2[0], x_2[1], color='blue', label='X2 (4,8)')
plt.plot([x_1[0], x_2[0]], [x_1[1], x_2[1]], linestyle='--', color='black')

# Añade la etiqueta de la distancia euclidiana
plt.text((x_1[0] + x_2[0]) / 2, (x_1[1] + x_2[1]) / 2, f"Distancia Euclidiana: {round(distancia_euclidiana, 2)}", ha='center', va='center')
plt.text(6.5, 7.8, f"Distancia Mahatan: {round(distancia_manhatan, 2)}", ha='center', va='center')
plt.text(4, 2, f"Distancia Maxima: {round(distancia_maxima, 2)}", ha='center', va='center')


# Dibujar las líneas de la distancia de Manhattan
plt.plot([x_1[0], x_1[0]], [x_1[1], x_2[1]], color='red', linestyle='--')
plt.plot([x_1[0], x_2[0]], [x_2[1], x_2[1]], color='red', linestyle='--')


plt.xlabel('X')
plt.ylabel('Y')
plt.title('Gráfica de las tuplas X1 y X2')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Definimos las variables y agregamos nuestros datos
id = [1,2,3,4,5,6,7,8]
color_ojos = ['Café', 'Negro', 'Azul', 'Negro', 'Café', 'Verde', 'Negro', 'Café']
sexo = ['F', 'M', 'F', 'M', 'F', 'M', 'M', 'M']
edad = [25, 22, 29, 26, 23, 29, 27, 25]

data = pd.DataFrame({'id': id, 'color_ojos': color_ojos, 'sexo': sexo, 'edad': edad})
data
```

* Distancia euclidiana:

$d(x_1,x_4) = \sqrt{ \sum{(1^2, 0^2, -1^2)}} =  1.41 $

$d(x_2,x_8) = \sqrt{ \sum{(1^2, 0^2, -3^2)}} =  3.16 $

* Distancia Manhatan:

$d(x_1,x_4) = \sum{|1, 0, -1|} =  0 $

$d(x_2,x_8) = \sum{|1, 0, -3|} =  -2 $


Para la normalización de los datos de edad se utiliza la soguiente formula a cada uno de los datos:

$$
col[A_j]^{'} = \frac{col[A_j] - min_{A_j}}{max_{A_j}-min_{A_j}}
$$

```python
# normalizacion de los datos de edad
data['edad_norm'] = round((data['edad'] - data['edad'].min()) / (data['edad'].max() - data['edad'].min()),2)
data
```

### Ejercicio 3 Modelos

Con los datos anteriores:

* Calcula/genera la matriz de correlación de datos. Concluye tus resultados.

* Elaborar un modelo de regresión lineal. Concluye tus resultados

Al tener valores categoricos en las variables, realizamos la codificacion de los datos en variables numericas.

```python
# codificacion de los datos de sexo: 0 si es F y 1 si es M
data['sexo_norm'] = data['sexo'].map({'F': 0, 'M': 1})
# asignamos un valor numerico de acuerdo al valor de la etiqueta del color de ojos
data['color_ojos_norm'] = data['color_ojos'].map({'Café': 0, 'Negro': 1, 'Azul': 2, 'Verde': 3})
data
```

Matriz de correlación de los datos:
Con este conjunto de datos y acorde con los resultados obtenidos en la matriz de correlación de Pearson, se concluye:

* Edad y color de ojos: con un valor de 0.734, indica tener una correlación positiva moderadamente fuerte. Aunque esta relación debe interpretarse con cautela debido a la naturaleza arbitraria de la codificación del color de ojos.

* No hay una relación significativa entre edad y sexo, tampoco entre el sexo y el color de ojos.

```python
# Tamaño de la variable
fig = plt.subplots(1, sharex='col', figsize=(6, 3))
labels = ['edad_norm','sexo_norm','color_ojos_norm'] # etiqueta de las columnas
df_corr = data[['edad_norm','sexo_norm','color_ojos_norm']] # solo variables originales
# Calculo del coeficiente de correlacion
matrix = pd.DataFrame(df_corr.corr('pearson').values, index=labels, columns=labels)
# Mapa de calor
sns.heatmap(matrix, annot=True, fmt='.3f', cmap=sns.cubehelix_palette(rot=-.1))
plt.title('Coeficiente de correlación de Pearson')
plt.show()
```

Con la visualización de los datos de las tres variables que se consideran en este conjunto de datos, al igual que la matriz de correlación se observa un patron entre el color de los ojos y la edad.

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

xs = data['edad_norm']
ys = data['sexo_norm']
zs = data['color_ojos_norm']

ax.scatter(xs, ys, zs)

ax.set_xlabel('Edad Normalizada')
ax.set_ylabel('Sexo Normalizado')
ax.set_zlabel('Color de Ojos Normalizado')

plt.show()
```

Se obtiene el modelo de regresión lineal

```python
import statsmodels.api as sm # estimacion y validacion de regresion

#iniciamos el proceso de la determinacion del modelo de regresion lineal
X = pd.DataFrame({ 'beta0' : np.ones(data.shape[0])
                  ,'beta1' : data['color_ojos_norm']
                  ,'beta2' : data['sexo_norm']
                  })

# definimos la variable respuesta
y = data['edad_norm']

modelo = sm.OLS(y, X) # Especificamos el modelo a usar
modelo = modelo.fit() # Ajustamos el modelo
print(modelo.summary())
```

De acuerdo con los resultados obtenidos, se tiene el siguiente modelo:

$$edad = β0 + β1 * sexo + β2 * colorojos + \epsilon $$

$$edad = 0.3464 + 0.2654 sexo - 0.1229 colorojos + \epsilon$$

El modelo de regresión parece tener cierto poder explicativo, pero no es un modelo muy sólido. Los valores de R cuadrado y R cuadrado ajustado son moderados, la importancia general del modelo es cuestionable. Los coeficientes sugieren que la primera variable independiente podría tener un efecto significativo en "edad", pero la segunda variable independiente "sexo" no parece ser un predictor sólido.
