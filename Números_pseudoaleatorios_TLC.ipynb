### Teorema central del limite visualmente

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame()

for i in range(1, 50):
  muestra = np.random.uniform(1000, 0.6, 10)
  col = f"Muestra {i}"
  df[col] = muestra

df.head(5)
```

```python
df.mean()
```

```python
df_muestra_medias = pd.DataFrame(df.mean(), columns=["Muestra media"])
df_muestra_medias
```

```python
sns.distplot(df_muestra_medias)
```

```python
sns.histplot(df_muestra_medias)
```
